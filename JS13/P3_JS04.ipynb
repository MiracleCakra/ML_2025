# Praktikum 3
Praktikum 3 kali ini kita akan mencoba menggunakan Keras untuk Regresi, khususnya pada kasus Prediksi Harga Rumah.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf

# Contoh dataset (buat dummy data)
data = pd.DataFrame({
    'luas': [50, 60, 70, 80, 90],
    'harga': [500, 600, 700, 800, 900]
})

X = data[['luas']]
y = data[['harga']]

# Normalisasi
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = scaler.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=100)

# Evaluasi
print("Prediksi:", model.predict(X_test))
```

# Tugas 4:

1. Ubah learning rate.

2. Bandingkan hasil loss.

3. Praktikum berikut akan menggunakan data Boston untuk memprediksi harga rumah.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("TUGAS 4: ANALISIS PENGARUH LEARNING RATE TERHADAP PERFORMA MODEL")
print("Dataset: California Housing Price Prediction")
print("="*80)

# 1. Load Dataset
housing = fetch_california_housing()
X = housing.data
y = housing.target

print(f"\nInformasi Dataset:")
print(f"  - Jumlah sampel: {X.shape[0]}")
print(f"  - Jumlah fitur: {X.shape[1]}")
print(f"  - Target: Median house value ($100,000)")

# 2. Preprocessing
scaler = StandardScaler()
Xs = scaler.fit_transform(X)
X_train, X_val, y_train, y_val = train_test_split(Xs, y, test_size=0.2, random_state=42)

print(f"\nPembagian Data:")
print(f"  - Training set: {X_train.shape[0]} sampel")
print(f"  - Validation set: {X_val.shape[0]} sampel")

# 3. Konfigurasi Learning Rate untuk Eksperimen
lr_configs = [
    {"label": "LR = 0.0001", "value": 0.0001, "color": "#1f77b4"},
    {"label": "LR = 0.0005", "value": 0.0005, "color": "#2ca02c"},
    {"label": "LR = 0.001 (Default)", "value": 0.001, "color": "#ff7f0e"},
    {"label": "LR = 0.005", "value": 0.005, "color": "#d62728"},
    {"label": "LR = 0.01", "value": 0.01, "color": "#9467bd"}
]

print("\n" + "="*80)
print("EKSPERIMEN: Training Model dengan Berbagai Learning Rate")
print("="*80)

results_data = []
training_histories = []

for config in lr_configs:
    lr_label = config['label']
    lr_val = config['value']

    print(f"\n{'='*60}")
    print(f"Eksperimen: {lr_label}")
    print(f"{'='*60}")

    # Build model dengan arsitektur yang sama
    model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(32, activation='relu'),
        Dense(1)
    ])

    # Compile dengan learning rate berbeda
    model.compile(
        optimizer=Adam(learning_rate=lr_val),
        loss='mse',
        metrics=['mae']
    )

    # Training
    print("Training model...")
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=200,
        batch_size=32,
        verbose=0
    )

    # Evaluasi
    predictions = model.predict(X_val, verbose=0)

    # Hitung metrik
    train_loss_awal = history.history['loss'][0]
    train_loss_akhir = history.history['loss'][-1]
    val_loss_awal = history.history['val_loss'][0]
    val_loss_akhir = history.history['val_loss'][-1]

    train_mae_akhir = history.history['mae'][-1]
    val_mae_akhir = history.history['val_mae'][-1]

    rmse = np.sqrt(mean_squared_error(y_val, predictions))

    # Cek konvergensi
    penurunan_loss = ((train_loss_awal - train_loss_akhir) / train_loss_awal) * 100

    if np.isnan(val_loss_akhir) or val_loss_akhir > 100:
        convergence = " Tidak Konvergen"
    elif penurunan_loss > 90:
        convergence = " Konvergen Sempurna"
    elif penurunan_loss > 70:
        convergence = " Konvergen Baik"
    else:
        convergence = " Konvergen Lambat"

    # Output hasil
    print(f"  Loss Epoch Pertama  : {train_loss_awal:.4f}")
    print(f"  Loss Epoch Terakhir : {train_loss_akhir:.4f}")
    print(f"  Penurunan Loss      : {penurunan_loss:.2f}%")
    print(f"  Validation Loss     : {val_loss_akhir:.4f}")
    print(f"  RMSE                : {rmse:.4f}")
    print(f"  Status              : {convergence}")

    # Simpan hasil
    results_data.append({
        'Learning Rate': lr_label,
        'LR Value': lr_val,
        'Loss Awal': round(train_loss_awal, 4),
        'Loss Akhir': round(train_loss_akhir, 4),
        'Penurunan (%)': round(penurunan_loss, 2),
        'Val Loss': round(val_loss_akhir, 4),
        'Val MAE': round(val_mae_akhir, 4),
        'RMSE': round(rmse, 4),
        'Konvergensi': convergence
    })

    training_histories.append({
        'label': lr_label,
        'lr': lr_val,
        'history': history,
        'color': config['color']
    })

# Tampilkan tabel ringkasan
print("\n" + "="*80)
print("TABEL RINGKASAN HASIL EKSPERIMEN")
print("="*80)
df = pd.DataFrame(results_data)
print(df.to_string(index=False))

# Visualisasi Hasil
fig = plt.figure(figsize=(16, 10))

# Plot 1: Training Loss Comparison
ax1 = plt.subplot(2, 3, 1)
for hist in training_histories:
    if not np.isnan(hist['history'].history['loss'][-1]):
        ax1.plot(hist['history'].history['loss'],
                label=hist['label'],
                color=hist['color'],
                linewidth=2,
                alpha=0.8)
ax1.set_xlabel('Epoch', fontsize=11, fontweight='bold')
ax1.set_ylabel('MSE Loss', fontsize=11, fontweight='bold')
ax1.set_title('Training Loss', fontsize=12, fontweight='bold')
ax1.legend(fontsize=9)
ax1.grid(True, alpha=0.3)
ax1.set_ylim(bottom=0, top=5)

# Plot 2: Validation Loss Comparison
ax2 = plt.subplot(2, 3, 2)
for hist in training_histories:
    if not np.isnan(hist['history'].history['val_loss'][-1]):
        ax2.plot(hist['history'].history['val_loss'],
                label=hist['label'],
                color=hist['color'],
                linewidth=2,
                alpha=0.8)
ax2.set_xlabel('Epoch', fontsize=11, fontweight='bold')
ax2.set_ylabel('MSE Loss', fontsize=11, fontweight='bold')
ax2.set_title('Validation Loss', fontsize=12, fontweight='bold')
ax2.legend(fontsize=9)
ax2.grid(True, alpha=0.3)
ax2.set_ylim(bottom=0, top=5)

# Plot 3: MAE Comparison
ax3 = plt.subplot(2, 3, 3)
for hist in training_histories:
    if not np.isnan(hist['history'].history['val_mae'][-1]):
        ax3.plot(hist['history'].history['val_mae'],
                label=hist['label'],
                color=hist['color'],
                linewidth=2,
                alpha=0.8)
ax3.set_xlabel('Epoch', fontsize=11, fontweight='bold')
ax3.set_ylabel('MAE', fontsize=11, fontweight='bold')
ax3.set_title('Validation MAE', fontsize=12, fontweight='bold')
ax3.legend(fontsize=9)
ax3.grid(True, alpha=0.3)

# Plot 4: Bar Chart - Final Validation Loss
ax4 = plt.subplot(2, 3, 4)
valid_res = [r for r in results_data if not np.isnan(r['Val Loss']) and r['Val Loss'] < 10]
labels = [r['Learning Rate'] for r in valid_res]
val_losses = [r['Val Loss'] for r in valid_res]
colors = [h['color'] for h in training_histories if h['label'] in labels]
ax4.bar(range(len(labels)), val_losses, color=colors, alpha=0.7, edgecolor='black')
ax4.set_xticks(range(len(labels)))
ax4.set_xticklabels(labels, rotation=15, ha='right', fontsize=9)
ax4.set_ylabel('Final Val Loss', fontsize=11, fontweight='bold')
ax4.set_title('Perbandingan Final Validation Loss', fontsize=12, fontweight='bold')
ax4.grid(axis='y', alpha=0.3)

# Plot 5: Bar Chart - RMSE
ax5 = plt.subplot(2, 3, 5)
rmse_vals = [r['RMSE'] for r in valid_res]
ax5.bar(range(len(labels)), rmse_vals, color=colors, alpha=0.7, edgecolor='black')
ax5.set_xticks(range(len(labels)))
ax5.set_xticklabels(labels, rotation=15, ha='right', fontsize=9)
ax5.set_ylabel('RMSE', fontsize=11, fontweight='bold')
ax5.set_title('Perbandingan RMSE', fontsize=12, fontweight='bold')
ax5.grid(axis='y', alpha=0.3)

# Plot 6: Learning Curve (Best Model)
ax6 = plt.subplot(2, 3, 6)
best_idx = np.argmin([r['Val Loss'] for r in valid_res])
best_hist = [h for h in training_histories if h['label'] == valid_res[best_idx]['Learning Rate']][0]
ax6.plot(best_hist['history'].history['loss'], label='Train Loss', color='blue', linewidth=2)
ax6.plot(best_hist['history'].history['val_loss'], label='Val Loss', color='red', linewidth=2)
ax6.set_xlabel('Epoch', fontsize=11, fontweight='bold')
ax6.set_ylabel('MSE Loss', fontsize=11, fontweight='bold')
ax6.set_title(f'Learning Curve Terbaik: {best_hist["label"]}', fontsize=12, fontweight='bold')
ax6.legend(fontsize=9)
ax6.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*80)
print("ANALISIS DAN KESIMPULAN")
print("="*80)

# Temukan model terbaik
valid_results = [r for r in results_data if not np.isnan(r['Val Loss']) and r['Val Loss'] < 10]
best_model = min(valid_results, key=lambda x: x['Val Loss'])

print(f"\n MODEL TERBAIK:")
print(f"   Learning Rate: {best_model['Learning Rate']}")
print(f"   Validation Loss: {best_model['Val Loss']:.4f}")
print(f"   RMSE: {best_model['RMSE']:.4f}")
print(f"   Penurunan Loss: {best_model['Penurunan (%)']:.2f}%")

print(f"\n KARAKTERISTIK LEARNING RATE:")

# Analisis LR kecil
lr_kecil = [r for r in results_data if r['LR Value'] <= 0.0005]
if lr_kecil:
    avg_penurunan = np.mean([r['Penurunan (%)'] for r in lr_kecil])
    print(f"\n   Learning Rate KECIL (≤0.0005):")
    print(f"   • Penurunan loss rata-rata: {avg_penurunan:.2f}%")

# Analisis LR sedang
lr_sedang = [r for r in results_data if 0.0005 < r['LR Value'] <= 0.005]
if lr_sedang:
    avg_loss = np.mean([r['Val Loss'] for r in lr_sedang if not np.isnan(r['Val Loss'])])
    print(f"\n   Learning Rate SEDANG (0.0005-0.005):")
    print(f"   • Validation loss rata-rata: {avg_loss:.4f}")
    print(f"   • Karakteristik: Keseimbangan antara kecepatan dan stabilitas")
    print(f"   • Cocok untuk: Sebagian besar kasus praktis")

# Analisis LR besar
lr_besar = [r for r in results_data if r['LR Value'] > 0.005]
if lr_besar:
    print(f"\n   Learning Rate BESAR (>0.005):")
    gagal = sum(1 for r in lr_besar if '❌' in r['Konvergensi'])

print(f"\n REKOMENDASI:")
print(f"   Untuk dataset California Housing dengan arsitektur [64, 32, 1]:")
print(f"   • Learning Rate optimal: {best_model['LR Value']}")
print(f"   • Menghasilkan konvergensi yang baik dengan performa terbaik")
print(f"   • Validation loss stabil dan RMSE minimal")

print("\n" + "="*80)
```
