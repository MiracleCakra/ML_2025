# Tugas Praktikum
Gunakan JST untuk klasifikasi angka tulisan tangan (MNIST).

Langkah:

Load dataset MNIST dari Keras.

Bangun model dengan 2 hidden layer.

Latih model dan evaluasi akurasi.

```python
import tensorflow as tf
import time # Import library time untuk mengukur durasi pelatihan
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# --- 1. Load & Preprocess Data ---
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 -> 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# --- 2. Bangun Model JST ---
model = Sequential([
    # Input Layer: Ubah gambar 28x28 menjadi vektor 1D (784 fitur)
    Flatten(input_shape=(28, 28)),

    # Hidden Layer 1: 128 neuron, aktivasi ReLU
    Dense(128, activation='relu'),

    # Hidden Layer 2: 64 neuron, aktivasi ReLU
    Dense(64, activation='relu'),

    # Output Layer: 10 kelas (angka 0-9), aktivasi Softmax (wajib untuk klasifikasi multikelas)
    Dense(10, activation='softmax')
])

# --- 3. Kompilasi Model ---
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# --- 4. Latih Model ---
print("Mulai pelatihan...")
start_time = time.time() # Mulai stopwatch

# Epochs=10 biasanya cukup untuk MNIST mencapai akurasi >97%
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

end_time = time.time() # Hentikan stopwatch
print(f"Waktu pelatihan: {end_time - start_time:.2f} detik")

# --- 5. Evaluasi Model ---
print("\nEvaluasi pada data uji:")
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi Final: {acc:.4f}")
```

## C. Eksperimen Aktivasi: Sigmoid vs ReLU

```python
import tensorflow as tf
import time
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# --- 1. Load & Preprocess Data ---
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 -> 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# --- 2. Bangun Model JST ---
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='sigmoid'),
    Dense(64, activation='sigmoid'),
    Dense(10, activation='softmax')
])

# --- 3. Kompilasi Model ---
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# --- 4. Latih Model ---
print("Mulai pelatihan...")
start_time = time.time() # Mulai stopwatch

history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

end_time = time.time() # Hentikan stopwatch
print(f"Waktu pelatihan: {end_time - start_time:.2f} detik")

# --- 5. Evaluasi Model ---
print("\nEvaluasi pada data uji:")
loss, acc = model.evaluate(X_test, y_test)
print(f"Akurasi Final: {acc:.4f}")
```

# Laporan Analisis Praktikum: Variasi Arsitektur JST pada MNIST

### 1. Ringkasan Hasil Eksperimen

| Skenario | Konfigurasi Model | Waktu (Detik) | Akurasi Final | Loss Final |
| :--- | :--- | :--- | :--- | :--- |

---

### 2. Analisis Mendalam

#### **A. Analisis Akurasi & Model Terbaik (Skenario A)**
* **Hasil:** Skenario A (Kode A) menjadi pemenang akurasi dengan **97.82%**.
* **Penyebab Teknis:** Kode A menggunakan jumlah neuron yang paling besar (**256** di layer pertama dan **128** di layer kedua).
* **Penjelasan:** Semakin banyak neuron, semakin banyak "kapasitas memori" (parameter) yang dimiliki jaringan untuk menangkap detail pola tulisan tangan yang rumit. Matriks bobot yang besar ini memungkinkan model membedakan variasi angka dengan lebih presisi dibandingkan model yang lebih kecil.

#### **B. Analisis "Deep Learning" yang Kurang Optimal (Skenario B)**
* **Hasil:** Skenario B (Kode B) justru memiliki akurasi terendah (**97.29%**) dan Loss tertinggi (**0.1467**), meskipun memiliki 3 hidden layer.
* **Penyebab Teknis:** Menambah kedalaman (*depth*) pada data sederhana seperti MNIST tidak selalu menjamin hasil lebih baik.
* **Penjelasan:** Kemungkinan terjadi **Overfitting** atau degradasi sinyal. Dengan arsitektur yang lebih dalam (128->64->32), model mungkin menjadi terlalu kompleks untuk data yang sebenarnya sederhana, sehingga "menghafal" noise pada data latih dan kesulitan saat diuji (terlihat dari *Loss* validasi yang tinggi: 0.14).

#### **C. Fenomena Unik Fungsi Aktivasi Sigmoid (Skenario C)**
* **Hasil:** Skenario C (Kode C) menempati posisi kedua dalam akurasi (**97.69%**) namun memiliki **Loss terendah (0.0857)**.
* **Analisis "Slow Start":**
    * Perhatikan Epoch 1 pada hasil C (Gambar 3ce7e0): Akurasi hanya **76.36%** dan Loss **0.94**. Ini jauh lebih buruk dibanding ReLU (A dan B) yang langsung start di atas 85%.
    * Ini disebabkan oleh karakteristik **Sigmoid** yang rentan terhadap *Vanishing Gradient* di awal, membuat pembelajaran lambat.
* **Analisis "Strong Finish":**
    * Meskipun lambat di awal, model ini berhasil mengejar ketertinggalan dan menghasilkan prediksi yang sangat stabil (*Loss* kecil) di akhir. Ini menunjukkan bahwa untuk jaringan yang tidak terlalu dalam (hanya 2 layer), Sigmoid masih cukup kompetitif.

#### **D. Analisis Efisiensi Waktu (Computational Cost)**
1.  **Terlambat (Skenario A - 102.67s):**
    Ini wajar karena jumlah parameter yang harus dihitung sangat masif. Layer pertama saja (784 input ke 256 neuron) memiliki bobot 2x lipat lebih banyak dibanding skenario B atau C.
2.  **Tercepat (Skenario C - 82.91s):**
    Skenario C paling cepat karena strukturnya sederhana (hanya 2 layer: 128, 64) dibandingkan B yang punya 3 layer, dan jumlah neuronnya lebih sedikit dibandingkan A.
